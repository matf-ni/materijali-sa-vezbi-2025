# Raznovrsnost spektrograma

Ova sveska približava jedinstvenost audio spektrograma. 

Možete eksperimentisati sa različitim zvučnim zapisima različitih instrumenata (`guitar.wav`, `bass_guitar.wav`, `drum.wav`, `piano.wav` i `organ.wav`) i uporediti dobijene spektrograme. Uzorci su preuzeti sa [MusicRadar](https://www.musicradar.com/) sajta.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from scipy import signal
from scipy.io import wavfile
from IPython import display
```

### Učitavanje i analiza zvučnog uzorka

```python
audio_path = 'data/bass_guitar.wav'
```

```python
display.Audio(audio_path)
```

```python
Fs, samples = wavfile.read(audio_path)
```

```python
print('Frekvencija uzorkovanja: {} Hz'.format(Fs))
```

```python
print('Broj uzoraka: ', samples.shape[0])
```

Neki uzorci su jednokanalni, a neki dvokanalni.

```python
number_of_channels = samples.ndim
```

```python
print('Broj kanala: ', number_of_channels)
```

Ako je neophodno, vršićemo i uprosečavanje vrednosti uzoraka sa očitanih kanala.

```python
if number_of_channels == 2:
    samples = samples.mean(axis=1)
```

### Spektrogram

```python
f, t, s = signal.spectrogram(samples, Fs)
```

```python
plt.pcolormesh(t, f, np.log(s), shading='auto')
plt.colorbar()
plt.xlabel('Time [sec]')
plt.ylim(0, 2500)
plt.ylabel('Frequency [Hz]')
plt.show()
```
